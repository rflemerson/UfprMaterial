```python
import numpy as np
import matplotlib.pyplot as plt
```

1) Utilizando sen, cos e linspace do NumPy mais matplotlib, calcule e apresente o gráfico de sen(x) e de cos(x) no intervalo \[-PI, +PI\] no mesmo plot (pegando PI do NumPy). Use plot e show de forma básica.

```python
x = np.linspace(-np.pi, np.pi, 100)

f1 = np.sin(x)
f2 = np.cos(x)

plt.plot(x, f1, 'b-', label='f(x) = sen(x)')
plt.plot(x, f2, 'r-', label='g(x) = cos(x)')

plt.show()
```

2) Utilizando o mesmo intervalo, use agora o cos(x) como abscissa e o sen(x) como ordenada. Pesquise a função sobre axes no matplotlib para deixar o gráfico quadrado, mesma proporção largura x altura, caso contrário o círculo parecerá uma elipse.

```python
x = np.linspace(-np.pi, np.pi, 100)

f1 = np.sin(x)
f2 = np.cos(x)

plt.plot(f2, f1, 'b-')
plt.axis("scaled")

plt.show()
```

3) Empregando sen(x) no mesmo intervalo, mas agora usando np.arange (com passo 0.2), plote 6 gráficos, de sen(x) até sen(x-1), deslocando cada plot de 0.2, ou sejam sen(x), sen(x-0.2) etc., usando para chamada ao plot os seguintes modelos de linhas: '-'    '--'    ':'     '-.'      '.'      'o'

```python
x = np.linspace(-np.pi, np.pi, 100)
stl = ['g-', 'b--', 'y:', 'g-.', 'g.', 'ro']

for i, s in zip(np.arange(0, 1.1, 0.2), stl):
    f1 = np.sin(x-i)
    plt.plot(x, f1, s)
```

4) Plote a curva dada por:

   x = sen t

   y = 2 sen 4t
   
   Com t no intervalo \[0, 2PI\] com 100 pontos.

```python
t = np.linspace(0, 2 * np.pi, 100)

x = np.sin(t)
y = 2 * np.sin(4 * t)

plt.plot(t, x, 'b-')
plt.plot(t, y, 'r-')
```

5) Plote o polinômio:

   f(x) = 4xˆ5 -12xˆ3 -6x^2 -x -2

   Com x no intervalo \[-2, 2\] com 200 pontos. 

```python
def f(x): return 4*(x**5) - 12*(x**3) - 6*(x**2) - x - 2

x = np.linspace(-2, 2, 200)

y = f(x)

plt.plot(x, y, 'b-')
plt.grid()
```
